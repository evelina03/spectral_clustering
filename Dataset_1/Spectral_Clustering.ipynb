```python
import numpy as np
# formatting the float to be three digit 
float_formatter = lambda x: "%.3f" % x
import pandas as pd
np.set_printoptions(formatter={'float_kind':float_formatter})
from sklearn.datasets import make_circles
from sklearn.cluster import SpectralClustering, KMeans
from sklearn.metrics import pairwise_distances
from matplotlib import pyplot as plt
import networkx as nx
import seaborn as sns
sns.set()
```

```python
df = pd.read_csv("HeartFailure.csv")
df.head()
```

```python
df_cluster = df.iloc[:,1:13]
```

```python
G = pairwise_distances(df_cluster, metric="euclidean")
G

```

```python
def draw_graph(G):
    pos = nx.spring_layout(G)
    nx.draw_networkx_nodes(G, pos)
    nx.draw_networkx_labels(G, pos)
    nx.draw_networkx_edges(G, pos, width=1.0, alpha=0.5)
```

```python
Graph=nx.from_numpy_matrix(G)
draw_graph(Graph)
```

```python
W = nx.adjacency_matrix(Graph)
print(W.todense())
```

```python
# degree matrix
D = Graph.degree()
D
```

```python
degrees = [val for (node, val) in Graph.degree()]
# degree matrix
D = np.zeros((len(degrees), len(degrees)))
np.fill_diagonal(D, degrees)
print('degree matrix:')
print(D)
```

```python
# laplacian matrix
L = D - W
print('laplacian matrix:')
print(L)
```

```python
e, v = np.linalg.eig(L)
# eigenvalues
print('eigenvalues:')
print(e)
# eigenvectors
print('eigenvectors:')
print(v)
```

```python
fig = plt.figure()
ax1 = plt.subplot(121)
plt.plot(e)
ax1.title.set_text('eigenvalues')
i = np.where(e < 10e-6)[0]
ax2 = plt.subplot(122)
plt.plot(v[:, i[0]])
fig.tight_layout()
plt.show()
```

```python
U = np.array(v[:, i[0]])
km = KMeans(init='k-means++', n_clusters=3)
df['clusters'] = km.fit_predict(U)
df
```
